С OpenCV

```python
import pytesseract
import cv2
import numpy as np
image = cv2.imread("test3.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
text = pytesseract.image_to_string(gray, lang='rus')
print(text)
```

С Pillow

```python
import pytesseract
from PIL import Image

# Загрузка изображения с текстом
image = Image.open("test2.jpg")

# Распознавание текста
string = pytesseract.image_to_string(image, lang='rus')

# Вывод распознанного текста в консоль
print(string)
```

дополнительно - вывод изображения с помощью tf и plt

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, array_to_img

#tf.enable_eager_execution()

img = load_img("test1.jpg")
img = tf.convert_to_tensor(np.asarray(img))
image = tf.image.resize(img,(400, 400))
to_img = array_to_img(image)
to_img.show()
```

```python
import matplotlib.pyplot as plt
plt.imshow(to_img)
plt.show()
```
